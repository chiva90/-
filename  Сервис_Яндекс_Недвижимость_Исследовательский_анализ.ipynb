```python
df['public_year'] = df['first_day_exposition'].dt.year   # –ì–æ–¥ –ø—É–±–ª–∏–∫–∞—Ü–∏–∏
df['public_month'] = df['first_day_exposition'].dt.month # –ú–µ—Å—è—Ü –ø—É–±–ª–∏–∫–∞—Ü–∏–∏
df['public_day'] = df['first_day_exposition'].dt.weekday # –î–µ–Ω—å –ø—É–±–ª–∏–∫–∞—Ü–∏–∏
```

```python
def public_day(row):
    list_days = {0: '–ü–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫', 1: '–í—Ç–æ—Ä–Ω–∏–∫', 2: '–°—Ä–µ–¥–∞', 3: '–ß–µ—Ç–≤–µ—Ä–≥', 4: '–ü—è—Ç–Ω–∏—Ü–∞', 5: '–°—É–±–±–æ—Ç–∞', 6: '–í–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ'}
    days = row['public_day'] 
    if days == 0:            
        days = list_days[0]  
        return days          
    elif days == 1:         
        days = list_days[1]
        return days        
    elif days == 2:
        days = list_days[2]
        return days      
    elif days == 3:
        days = list_days[3]
        return days      
    elif days == 4:
        days = list_days[4]
        return days
    elif days == 5:
        days = list_days[5]
        return days  
    elif days == 6:
        days = list_days[6]
        return days
    else:
        return days
```

```python
df['public_day'] = df.apply(public_day, axis=1)
```

```python
def public_month(row):
    list_month = {1: '–Ø–Ω–≤–∞—Ä—å', 2: '–§–µ–≤—Ä–∞–ª—å', 3: '–ú–∞—Ä—Ç', 4: '–ê–ø—Ä–µ–ª—å', 5: '–ú–∞–π', 6: '–ò—é–Ω—å', 7: '–ò—é–ª—å', 8: '–ê–≤–≥—É—Å—Ç', 9: '–°–µ–Ω—Ç—è–±—Ä—å', 10: '–û–∫—Ç—è–±—Ä—å', 11: '–ù–æ—è–±—Ä—å', 12: '–î–µ–∫–∞–±—Ä—å'}
    month = row['public_month'] 
    if month == 1:              
        month = list_month[1]  
        return month            
    elif month == 2:            
        month = list_month[2]  
        return month      
    elif month == 3:
        month = list_month[3]
        return month      
    elif month == 4:
        month = list_month[4]
        return month
    elif month == 5:
        month = list_month[5]
        return month  
    elif month == 6:
        month = list_month[6]
        return month
    elif month == 7:
        month = list_month[7]
        return month 
    elif month == 8:
        month = list_month[8]
        return month    
    elif month == 9:
        month = list_month[9]
        return month    
    elif month == 10:
        month = list_month[10]
        return month
    elif month == 11:
        month = list_month[11]
        return month
    elif month == 12:
        month = list_month[12]
        return month    
    else:
        return month
```

```python
df['public_month'] = df.apply(public_month, axis=1)
```

* –¢–∏–ø –ø–ª–æ—â–∞–¥–∏:
* –û—á–µ–Ω—å –º–∞–ª–µ–Ω—å–∫–∞—è: 0-20%
* –ú–∞–ª–µ–Ω—å–∫–∞—è: 21-40%
* –°—Ä–µ–¥–Ω—è—è: 41-60%
* –ë–æ–ª—å—à–∞—è: 61-80%

```python
df['total_area'].quantile([.20, .40, .60, .80]) # —Å–¥–µ–ª–∞–µ–º —Ä–∞–∑–±–∏–≤–∫—É —Å —à–∞–≥–æ–≤ 20%
```

```python
def area_type(row):       
    area = row['total_area'] 
    if area <= 38:            
        return '–û—á–µ–Ω—å –º–∞–ª–µ–Ω—å–∫–∞—è'  
    if area <= 46 and area > 38: 
        return '–ú–∞–ª–µ–Ω—å–∫–∞—è'
    if area <= 58 and area > 46:
        return '–°—Ä–µ–¥–Ω—è—è'
    if area <= 75 and area > 58:
        return '–ë–æ–ª—å—à–∞—è'
    if area > 75:
        return '–û—á–µ–Ω—å –±–æ–ª—å—à–∞—è'
    else:                                
        return '–ù–µ—Ç –¥–∞–Ω–Ω—ã—Ö'              
    
df['area_type'] = df.apply(area_type, axis=1)
```

```python
def price_type(row):       
    price = row['last_price'] 
    if price <= 3085:            
        return '–û—á–µ–Ω—å –Ω–∏–∑–∫–∞—è —Ü–µ–Ω–∞'  
    if price <= 3900 and price > 3085: 
        return '–ù–∏–∑–∫–∞—è —Ü–µ–Ω–∞'
    if price <=  4800 and price > 3900:
        return '–°—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞'
    if price <=  6300 and price >  4800:
        return '–í—ã—Å–æ–∫–∞—è —Ü–µ–Ω–∞'
    if price > 6300:
        return '–û—á–µ–Ω—å –≤—ã—Å–æ–∫–∞—è —Ü–µ–Ω–∞'
    else:                                
        return '–ù–µ—Ç –¥–∞–Ω–Ω—ã—Ö'              
    
price_new['price_type'] = df.apply(price_type, axis=1)
```
